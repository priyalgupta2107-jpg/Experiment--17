```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2,3,2.5,4,3.5],
    'Marks': [50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours':[6,7,6,8,7],
    'Assignments_Completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1.Line Chart(Trend)

plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.title("priyal")
plt.xlabel("Days")
plt.ylabel("Study Hours")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Advanced Line Chart
plt.figure(figsize=(7,4))

plt.plot(df['Days'], df['Study_Hours'],
         marker='o',color='blue',label='Study Hours')

plt.plot(df['Days'], df['Marks'],
         marker='s',color='red',label='Marks')

plt.title("priyal")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#2. Bar Chart(Comparision)
plt.bar(df['Days'],df['Marks'],color='green')
plt.title("priyal")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.show()
```

```python
#Advanced Bar Graph
bars =plt.bar(df['Days'],df['Marks'],color='red')
for bar in bars:
   y = bar.get_height()
   plt.text(bar.get_x()+bar.get_width()/2,y,str(y),ha='center',va='bottom')

   plt.title("priyal")
   plt.xlabel("Days")
   plt.ylabel("Marks")
   plt.grid(axis='y') # adds horizontal grid for readability
plt.show()
```

```python
#3. Histogram(Distribution)
# Shows how values are spread
data ={
    'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
}
hist_df = pd.DataFrame(data)
plt.hist(hist_df['Marks'],bins=3, edgecolor='yellow',alpha=0.30)
plt.title("Priyal")
plt.xlabel("Marks")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Example data (more values for better histogram)
marks = [50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
# Calculate Mean
mean_value = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=6,color='skyblue',edgecolor='black')
# Mean Line
plt.axvline(mean_value,color='red',linestyle='--',linewidth=2,label='Mean')
plt.title("Priyal")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()

plt.grid(axis='y',linestyle='--',alpha=0.5)
display(mean_value)
plt.show()
```

```python
#4. Scatter Plot (Relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title("Priyal")
plt.xlabel("Study_Hours")
plt.ylabel("Marks")
plt.show()
```

```python
# Conditional Scatter Plot
# Example Category
df['Result'] = ['Fail','Pass','Fail','Pass','Pass']
colors =['red' if r=='Fail' else'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title("Priyal")
plt.xlabel("Study_Hours")
plt.ylabel("Marks")
plt.show()
```

```python
import pandas as pd

# Dataset-2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North','South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics','Food',
                 'Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Line Plot
sns.lineplot(x='Day',y='Sales',data=df)
plt.title("Priyal")
plt.show()
```

```python
# Bar Plot
sns.barplot(x='Day',y='Profit',data=df)
plt.title("Priyal")
plt.show()
```

```python
# Histogram
sns.histplot(df['Sales'],bins=5)
plt.title("Priyal")
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(x='Sales',y='Profit',data=df)
plt.title("Priyal")
plt.show()
```
